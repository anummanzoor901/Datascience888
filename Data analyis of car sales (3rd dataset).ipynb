### Imbalancing Data

To make my data imbalance I used random data creation method. 

website which I used to study about random sampling: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sample.html



# A) Low imbalance (65%)

first I will generate 65% imbalanced data and visualize it. 

```python
rows = df.sample(frac =.65)
rows
```

277472 rows are created randomly, and from the above display we can see that Nulls are randomly generated too. It created 65% random data. 

Next, I will combine and shuffle these randomly created rows with original dataset. 

```python
df = pd.concat([df, rows])
df
```

### Visualizing data

Vehicle purchase from 2000 to 2021

```python
plt.hist(list(df['Sales_in_thousands']), bins = 20)
plt.xlabel("Sales in thousands")
plt.ylabel("Frequency")
plt.title("Understanding the distribution of the variable - Sales_in_thousands")
plt.show()
```

### Top 10 'Manufacturers' with most 'Models' manufactured

```python
manufacturer_count = dict()
for each_manufacturer in list(df['Manufacturer']):
    if each_manufacturer not in manufacturer_count:
        manufacturer_count[each_manufacturer] = 1
    else:
        manufacturer_count[each_manufacturer] += 1

sorted_manufacturers = dict(sorted(manufacturer_count.items(), key = lambda item : item[1], reverse = True))

top_10_manufacturers = []
top_10_counts = []
for each_manufacturer in sorted_manufacturers:
    top_10_manufacturers.append(each_manufacturer)
    top_10_counts.append(sorted_manufacturers[each_manufacturer])
    if len(top_10_counts) == 10:
        break

fig = plt.figure(figsize = (10, 7))
plt.bar(top_10_manufacturers, top_10_counts, color = 'red', width = 0.4)
plt.xlabel("Top 10 Manufacturers")
plt.ylabel("Models manufactured")
plt.title("Top 10 manufacturers with most models manufactured")
plt.show()
```

We can know vehicle's price is increasing from median graph. Except for 2021, since the dataset was created in 2021, not all records are present. 

On observing the above graph, we can say that the manufacturers 'Dodge' and 'Ford' manufactured most models than any other manufacturer.

### Different types of vehicles used in the dataset

```python
vehicle_type_dict = dict()
for each_vehicle in df['Vehicle_type']:
    if each_vehicle not in vehicle_type_dict:
        vehicle_type_dict[each_vehicle] = 1
    else:
        vehicle_type_dict[each_vehicle] += 1

vehicle_type = list(vehicle_type_dict.keys())
vehicle_count = list(vehicle_type_dict.values())

plt.bar(vehicle_type, vehicle_count, color = 'red', width = 0.5)
plt.xlabel("Various Vehicle Types")
plt.ylabel("Count of each vehicle type")
plt.title("Bar graph showing the count of various vehicle types used in the dataset")
plt.show()
```

On observing the above graph, we can say that the count of passenger vehicle type is almost three times than car vehicle type
